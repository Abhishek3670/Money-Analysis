```python
import numpy as np
import pandas as pd
from difflib import SequenceMatcher

import os
```

```python
file_path = os.path.join(os.getcwd(), 'statements','JUL', 'transaction.xls')

# Read the XLS file with the first column as index
df = pd.read_excel(file_path, index_col=0)
```

```python
# Columns = ['Value Date', 'Transaction Date', 'Cheque Number','Transaction Remarks', 'Withdrawal Amount (INR )','Deposit Amount (INR )', 'Balance (INR )']
TRANS_DATE = df['Transaction Date']
TRANS_REMARKS = df['Transaction Remarks']
WITHDRAWAL_AMT = df['Withdrawal Amount (INR )']
DEPOSIT_AMT = df['Deposit Amount (INR )']
BALANCE = df['Balance (INR )']
```

```python
# Pre-proccess of Data
df = df.drop(['Value Date', 'Cheque Number'], axis=1)
df
```

```python
# Split the string by '/' and assign to new columns
# df[['Trasaction Type 1', 'Trasaction Type 2', 'Trasaction Type 3','Trasaction Type 4','Trasaction Type 5']] = df[TRANS_REMARKS].str.split('/', expand=True)

# Split the 'data' column, create a list of lists
TRANS_REMARKS = df["Transaction Remarks"]
data_list = [row.split('/') for row in df["Transaction Remarks"]]

# data_list
new_col_names = ['Sender(s) Trasaction Type', 'Sender(s) UTR', 'Recipient(s) Trasaction Type','Recipient(s) UTR','Bank', 'UniqueID']
index = pd.RangeIndex(start=1, stop=len(data_list) + 1)
df_new = pd.DataFrame.from_records(data_list, columns=new_col_names,index=index)  # Assuming all rows have the same structure
df = pd.concat([df, df_new], axis=1)  # Combine along columns
df.head()

```

```python
# Specify the column name and substring
substring = 'vaibhav'

# Filter for rows where 'substring' is found (case-sensitive) and handle missing values
filtered_df = df[df["Recipient(s) UTR"].notna() & df["Recipient(s) UTR"].str.contains(substring)]

# Print the filtered DataFrame
print(filtered_df)
```

```python
import os
import datetime

# Month abbreviations dictionary
month_abbr = {
    1: "JAN",
    2: "FEB",
    3: "MAR",
    4: "APR",
    5: "MAY",
    6: "JUN",
    7: "JUL",
    8: "AUG",
    9: "SEP",
    10: "OCT",
    11: "NOV",
    12: "DEC"
}

# Get current date
today = datetime.date.today()

# Create folders for all months (up to current month)
for month_num in range(1, today.month + 1):
  month_str = month_abbr[month_num]
  folder_path = os.path.join("statements", month_str)

  # Check if folder already exists (optional)
  if not os.path.exists(folder_path):  # Uncomment for checking existence

    # Create the folder
    os.makedirs(folder_path, exist_ok=True)  # Handles existing folders
  print(f"Created folder: {folder_path}")

print("All folders created (up to current month).")
```
